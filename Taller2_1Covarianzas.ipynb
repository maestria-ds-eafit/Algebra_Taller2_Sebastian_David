```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## A) Body Fat Prediction Dataset

```python
df=pd.read_csv("bodyfat.csv")
df.head()
```

## Selecting 4 continous variables

```python
df=df[["BodyFat","Weight","Height","Abdomen"]]
df.head()
```

## B) Creating a pairplot for each couple of variables

```python
sns.pairplot(df)
plt.show()
```

## Calculating means vector

```python
variable_means=df.mean()
variable_means
```

```python

# Create a pair plot for all variables
sns.set(style="ticks")
g = sns.pairplot(df)

# Add points representing means of each variable pair
for i, ax_row in enumerate(g.axes):
    for j, ax in enumerate(ax_row):
        if i != j:
            ax.scatter(variable_means[j], variable_means[i], color='red', marker='o', s=100)

        
plt.show()
```

## Calculating Covariance Matrix

```python
df.describe()
```

```python
covariance_matrix = df.cov()

# Print the covariance matrix
print(covariance_matrix)
```

## Recentering data

```python
z = df - variable_means
z
```
